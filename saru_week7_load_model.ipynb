## **Loading mnist_model and predicting a MNIST image from th URL**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

## **Function to download and plot images from user's url**

```python
import urllib.request
from matplotlib import image

def picture(url):
    urllib.request.urlretrieve(url,'mnist_image.jpg')
    
    img = image.imread('mnist_image.jpg')
    plt.figure(figsize=(10,10))
    plt.imshow(img)
    plt.axis(False)
    return img
```

```python
url = input("Enter URL of MNIST image: ")
```

```python
myImage = picture(url)
print(myImage.shape)
```

## **Loading the model**

```python
loaded_model = keras.models.load_model("mnist_model")
```

## **Resize our image**

```python
import cv2
import numpy as np

img_array = np.asarray(myImage)
resized = cv2.resize(img_array, (28, 28 ))
gray_scale = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY) #(28, 28)
image = cv2.bitwise_not(gray_scale)

image = image / 255
image = image.reshape(1, 784)
print(image.shape)
```

## **Make a prediction**

```python
prediction = loaded_model.predict_classes(image)
print(f'Your predicted number is: {prediction}')
```
